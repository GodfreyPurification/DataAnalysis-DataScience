```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV file into a pandas DataFrame
df = pd.read_csv("StudentsPerformance.csv")

# Extract scores for each subject
math_scores = df['math score']
reading_scores = df['reading score']
writing_scores = df['writing score']
```

```python
df
```

```python
# Set the range and bins
range_values = (0, 100)
bins = 100

# Plotting histograms
plt.figure(figsize=(12, 6))

plt.hist(math_scores, bins=bins, range=range_values, color='green', histtype='bar', rwidth=0.8, alpha=0.7, label='Math')
plt.hist(reading_scores, bins=bins, range=range_values, color='red', histtype='bar', rwidth=0.8, alpha=0.7, label='Reading')
plt.hist(writing_scores, bins=bins, range=range_values, color='yellow', histtype='bar', rwidth=0.8, alpha=0.7, label='Writing')

# Set labels and title
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Histogram Plot')
plt.legend(loc='upper right')

plt.show()
```
